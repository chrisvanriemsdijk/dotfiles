```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
map = gpd.read_file("Map/cb_2018_us_state_500k.shp")
crs = {'init': 'epsg:4326'}
data = pd.read_csv("sample_data/all_sites.csv")
points = [Point(xy) for xy in zip(data["LAT"], data["LNG"])]
geo_df = gpd.GeoDataFrame(data,  geometry = points)
fig, ax = plt.subplots(figsize=(15,15))
map.plot(ax = ax, alpha=0.4, color="grey")
geo_df.plot(ax = ax, markersize=20, color="blue", marker = "o")

```
